```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
boston = fetch_california_housing()
x = boston.data
y = boston.target
```

```python
scaler = StandardScaler().fit(x)
x = scaler.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=x.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=15, batch_size=8, verbose=1)
```

```python
mse = model.evaluate(x_test, y_test, verbose=0)
print("Mean Squared Error : ", mse)
```

```python
y_pred = model.predict(x_test)
print("Predicted values :\n", y_pred[:5])
print("True values :\n", y_test[:5])
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(y_test[:50], label="True Values", color='blue', marker='o')
plt.plot(y_pred[:50], label="Predicted Values", color='red', marker='x')
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("True vs Predicted Values")
plt.legend()
plt.show()
```
